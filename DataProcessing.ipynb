### October 12

```python
October12 = ObservationNight()
October12.loadFiles('Data/Calibrated/rx-180.000secs.Light.Open.000000%s.fit',2,1,39)
for i in range(4):
    October12.popFile(-1) # frames too cloudy to see asteroid
```

I found these reference stars and their magnitudes by searching my field of view in Vizier and then visually matching the results to my image.

```python
October12.newRefStar('C1',230,856,12.2) # 4 12 02 +23 03 20
October12.newRefStar('C2',257,747,13.01) # 4 12 04 +23 01 08
October12.newRefStar('C3',611,272,13.92) # 4 12 34 +22 51 47
October12.setInitialAsteroid(441,438)
```

```python
October12.getFiles()[0].curveOfGrowth((230,856),30)
```

Choose an aperture size that's not near the asymptote to include as much signal from the targets as possible without reducing your SNR by including too much background. In this case, I chose 5 pixels. 6 or 7 probably would have been ok too.

```python
October12.setApertures(5)
```

```python
October12.graphStarRatios()
```

C2 is the worst so I'm taking it out.

```python
October12.delRefStar('C2')
```

```python
October12.graphAsteroidSignal()
```

The two outliers in this graph around 2.5 hours are from frames when it was cloudy.

```python
October12.graphAsteroidMags()
```

### October 17
